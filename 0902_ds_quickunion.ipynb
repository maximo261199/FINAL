```python
import graphviz as gv
```

```python
def drawDS(ds):
  graph = gv.Digraph("DisjointSet")
  graph.graph_attr['rankdir'] = "BT"
  for e, p in enumerate(ds):
    graph.node(str(e))
    if p != e:
      graph.edge(str(e), str(p))
  return graph
```

```python
ds = [0, 1, 2, 2, 8, 5, 5, 7, 8, 2]
drawDS(ds)
```

```python
def find(ds, e):
  print(ds[e])
  return ds[e]
```

```python
def union(ds, a, b):
  print(a)
  a = find(ds,  a)
  print(a)
  print(b)
  b = find(ds, b)
  print(b)
  if a == b: return

  for e, p in enumerate(ds):
    if p == a:
      ds[e] = b
```

```python
ds = [i for i in range(10)]
drawDS(ds)
```

```python
union(ds,  8,0)
drawDS(ds)
```
